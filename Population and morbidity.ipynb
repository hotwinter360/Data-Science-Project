```python
import requests 
import lxml.html as lh
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.macrotrends.net/countries/ISR/israel/population"
page = requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

```python
tbl=soup("table",attrs={"class":"historical_data_table"})[1]
```

```python
year=list()
pop=list()
growth_rate=list()

for row in tbl("tr"):
    cells=row("td")
    if(len(cells)<3):
        continue
    year.append(cells[0].get_text())  
    pop.append(cells[1].get_text())
    growth_rate.append(cells[2].get_text())
```

```python
df_pop=pd.DataFrame({"Year":year,"Population":pop,"Growth_rate":growth_rate})
df_pop.sort_values(by="Year",ascending=False,inplace=True)
df_rev = df_pop.head(13)
df_rev
```

```python
df_rev=df_rev.tail(9)
df_rev
```

```python
morbidity=[2764,2675,2759,2594,2646,2521,2382,2206,2268]
df_rev["morbidity"]=morbidity
df_rev
```

```python
df_rev.to_csv("pop and mor.csv")
```
