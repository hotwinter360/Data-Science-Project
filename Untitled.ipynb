```python
import pandas  as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_air=pd.read_csv("air stats.csv")
df_air.columns
```

```python
plt.plot(df_air.Year,df_air.pm25_avg,label='PM2.5 AVG')
plt.plot(df_air.Year,df_air.no2_avg,label='NO2 AVG')
plt.title("NO2 and PM2.5 avarage by years")
plt.xlabel("Years")
plt.ylabel("Average pollutant")
plt.legend()
```
