```python
import pandas  as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_air=pd.read_csv("air stats.csv")
df_air.columns
```

```python
plt.plot(df_air.Year,df_air.no2_avg ,label ='NO2 AVG')
plt.plot(df_air.Year,df_air.pm25_avg, label="PM2.5 AVG")
plt.title("Average over the years 2010 to 2018")
plt.xlabel("Year")
plt.ylabel("Annual average")
plt.legend()
plt.style.use('fivethirtyeight')
plt.savefig('Pollutants.png')
```

```python
df_mor=pd.read_csv("Morbidity.csv")
df_mor.columns
```

```python
plt.plot(df_mor.year,df_mor.population,c='r',label='growth rate')
plt.bar(df_mor.year,df_mor.population,label='Population')

plt.title("Population 2010 to 2018")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.style.use('fivethirtyeight')
plt.savefig("Population rise.png")
```

```python
plt.bar(df_mor.year,df_mor.morbidity, label='morbidity')
plt.title("morbidity 2010 to 2018")
plt.xlabel("Year")
plt.ylabel("morbidity")
plt.legend()
plt.style.use('fivethirtyeight')
plt.plot(df_mor.year,df_mor.morbidity,c="r", label='growth rate')
plt.savefig("morbidity.png")
```
