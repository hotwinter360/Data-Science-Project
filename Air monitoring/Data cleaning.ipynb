```python
import pandas as pd
import numpy as np
```

```python
def fill_zero (df):
    df.replace(to_replace="NoData",value=0,inplace=True)
    df    
```

```python
df_10=pd.read_csv("Air_monitoring_2010.csv")
df_11=pd.read_csv("Air_monitoring_2011.csv")
df_12=pd.read_csv("Air_monitoring_2012.csv")
df_13=pd.read_csv("Air_monitoring_2013.csv")
df_14=pd.read_csv("Air_monitoring_2014.csv")
df_15=pd.read_csv("Air_monitoring_2015.csv")
df_16=pd.read_csv("Air_monitoring_2016.csv")
df_17=pd.read_csv("Air_monitoring_2017.csv")
df_18=pd.read_csv("Air_monitoring_2018.csv")
df_12.head()
```

```python
fill_zero(df_10)
fill_zero(df_11)
fill_zero(df_12)
fill_zero(df_13)
fill_zero(df_14)
fill_zero(df_15)
fill_zero(df_16)
fill_zero(df_17)
fill_zero(df_18)
df_10.to_csv("Air_monitoring_2010.csv")
df_11.to_csv("Air_monitoring_2011.csv")
df_12.to_csv("Air_monitoring_2012.csv")
df_13.to_csv("Air_monitoring_2013.csv")
df_14.to_csv("Air_monitoring_2014.csv")
df_15.to_csv("Air_monitoring_2015.csv")
df_16.to_csv("Air_monitoring_2016.csv")
df_17.to_csv("Air_monitoring_2017.csv")
df_18.to_csv("Air_monitoring_2018.csv")
```
